# 2. Submit & monitor

Launches the jobs from `jobs.json` and lets you watch / cancel them. Run
`1_build.ipynb` first (and again whenever you change a notebook — see `3_sync`).

```python
import nb2slurm
wf, cfg = nb2slurm.load_config("control_config.json")
```

## Submit

`submit()` reads `jobs.json` by default — one job per leaf path.

```python
wf.submit(ssh=cfg, dry_run=True)     # preview the sbatch commands
job_ids = wf.submit(ssh=cfg)         # for real

# run a custom subset instead:
#   wf.submit([("NL", "123", "ssp126")], ssh=cfg)
```

## Monitor

Parsed `squeue` for your jobs.

```python
for row in wf.status(ssh=cfg):
    print(row)
```

## Cancel (if needed)

Cancels the jobs submitted in this session.

```python
wf.cancel(ssh=cfg)
```
